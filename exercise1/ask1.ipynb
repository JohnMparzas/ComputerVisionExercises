```python
import pandas as pd
import numpy as np
import scipy as sp
import matplotlib.image as mpimg
import skimage.data as fixtures
import matplotlib.pyplot as plt
from PIL import Image
from numpy.fft import fft2
from numpy import *
```

```python
A = np.array(Image.open('trikoupi6.png'))
plt.imshow(A, cmap="gray") # (returns an argument we can ignore)
plt.show()
```

### Εδω βλεπουμε την αρχικη εικονα

```python

A_gray = np.zeros((A.shape[0],A.shape[1]))
A=double(A)
if len(A.shape)==3:
    for i in range(A.shape[0]):
        for j in range(A.shape[1]):
            A_gray[i,j]=(A[i,j,0]+A[i,j,1]+A[i,j,2])/2
else:
      A_gray=np.copy(A)
    
    
```

```python
min_value=np.min(A_gray)
max_value=np.max(A_gray)
print('min_value : ',min_value)
print('max_value : ',max_value)
```

### Εδω βλεπουμε τη ελαχιστη και την μεγιστη τιμη των pixel της εικονας

```python
k=[60,100,150,200,260]



for k_i in k:
    A2=np.copy(A_gray)
    for i in range(A2.shape[0]):
        for j in range(A2.shape[1]): 
            if A2[i,j]>k_i:
                A2[i,j]=max_value
            else:
                A2[i,j]=min_value
                
    file_name='thresh_noteb(k='+str(k_i)+').png'
    A2 = np.uint8(A2)
    #Image.fromarray(A2).save(file_name)
                
    plt.imshow(A2, cmap="gray",label=str(k)) # (returns an argument we can ignore)
    plt.show()
```
